```python
import ccxt
import pprint
import pandas as pd
import numpy as np
from pandas import Series
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib as mpl
import time
```

```python
df = pd.read_csv("btc_day.csv")
df.drop(['Unnamed: 0','id','ranges','MA7','MA25','MA99','upperBB','lowerBB'], axis=1, inplace=True)
df.columns = ['날짜', '시가', '고가', '저가', '종가', '거래량']
df.set_index('날짜', inplace=True)
df
```

```python
def SMA(data, period, column='종가'):
    SMA = data[column].rolling(window=period).mean()
    df['SMA'] = SMA
    return data
```

```python
SMA(df, 20, '종가')
```

```python
df['상한선'] = df['SMA'] + (0.1 * df['SMA'])
df
```

```python
df['하한선'] = df['SMA'] - (0.1 * df['SMA'])
df
```

```python
pd.set_option('display.max_rows',None)
df
```

```python
df['시그널'] = df['종가'] > df['SMA']
df
```

```python
df['매도가'] = np.where(df['고가'] > df['상한선'], df['상한선'], 0)
df
```

```python
df['목표가'] = df['종가']
df
```

```python
# 초기설정
start = 10_000_000
start_money =10_000_000
high = start_money

max_bench = 0
mdd_bench = 0
    
df['전고점'] = 0
df = df.dropna()
    
# backtest
list1 = list(df['시그널'])
list2 = list(df['목표가'])
list3 = list(df['매도가'])
list4 = list(df['종가']) 
    
current_asset_list = []
asset_list = []
profit_list = []
profit_percent_list = []
maximum_list = []
mdd_list = []

for i,j in enumerate(list1):
    if j == True:
        possible_coin_number = start_money//list2[i]
        
        if list3[i] > 0:
            profit_percent = 100*((list3[i] - list2[i])*possible_coin_number)/start_money
            profit = (list3[i] - list2[i])*possible_coin_number
            
        current_asset = start_money
            
        asset = start_money + profit
            
        start_money = asset
            
        maximum = max(high,current_asset)
            
        high = maximum
    else:
        profit_percent = 0
        profit = 0
        asset = start_money
        current_asset = start_money
        maximum = max(high,current_asset)
        high = maximum
            
    current_asset_list.append(current_asset)
    asset_list.append(asset)
    profit_list.append(profit)
    maximum_list.append(maximum)
    profit_percent_list.append(profit_percent)
        
        
    if (asset - maximum) > 0:
        mdd = 0
    else:
        mdd = ((asset-maximum)/maximum)*100
            
            
    mdd_list.append(mdd)
    max_bench = max(max_bench,list4[i])
    mdd_bench = min(mdd_bench,((list4[i]-max_bench)/max_bench)*100)

df['자산'] = asset_list
df['기존자산'] = current_asset_list
df['전고점'] = maximum_list
df['수익률'] = profit_percent_list
df['수익금'] = profit_list
df['MDD'] = mdd_list
    
df['수익률'] = df['수익률']/100
df['누적수익률'] = (df['수익률']+1).cumprod()-1
df['수익률'] = df['수익률']*100
df['누적수익률'] = df['누적수익률']*100

list5 = list(df['자산'])
list6 = list(df['누적수익률'])

trade_day = int(len(df.index))
cagr = ((list5[-1]/list5[0])**(252/trade_day)-1)*100
cagr_bench = ((list4[-1]/list4[0])**(252/trade_day)-1)*100

# 주요 지표 구하기
backtest_list = {}
backtest_list['Start Money'] = start
backtest_list['Total returns'] = str(round(list6[-1],2)) + "%"
backtest_list['CAGR'] = str(round(cagr,2)) + "%"
backtest_list['CAGR benchmark'] = str(round(cagr_bench,2)) + "%"
backtest_list['MDD'] = str(round(min(mdd_list),2)) + "%"
backtest_list['MDD benchmark'] = str(round(mdd_bench,2)) + "%"
backtest_list['Periods'] = trade_day

result = pd.DataFrame(backtest_list,index=[1])
result = result.transpose()
result.columns = ['Algorithm1']
print("< Momentum Pinball 전략 > \n\n")
print(result)
```
